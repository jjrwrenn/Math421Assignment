```r
hmeq = read.csv(file = "C:\\Users\\student\\Documents\\Statistical Analysis using R\\Assignment 18\\hmeq.csv", header = TRUE)
```

```r
hmeq = hmeq[complete.cases(hmeq), ]
hmeq$BAD = as.factor(hmeq$BAD)
names(hmeq)[1] = 'target'

library(caret)
splitIndex <- createDataPartition(hmeq$target, p = 0.70, list = FALSE, times = 1)
train <- hmeq[ splitIndex,]
test  <- hmeq[-splitIndex,]

library(ranger)
model = ranger(target ~., data = train)
pred = predict(model, data = test)$predictions
cm = confusionMatrix(pred, test$target, positive = "1")

accuracy = cm[[3]][1]
balancedAccuracy = cm[[4]][11]

accuracy
balancedAccuracy
```

```r
train0 = train[train$target == '0',]
train1 = train[train$target == '1',]

train00 = train0[sample(1:nrow(train0), nrow(train1)),]
train_under <- rbind(train00, train1)

model = ranger(target ~., data = train_under)
pred = predict(model, data = test)$predictions
cm = confusionMatrix(pred, test$target, positive = "1")

accuracy00 = cm[[3]][1]
balancedAccuracy00 = cm[[4]][11]

accuracy00
balancedAccuracy00
```

```r
Quick_bal = function (dataSet, OU) {
  library("caret")
  splitIndex <- createDataPartition(dataSet$target, p = 0.70, list = FALSE, times = 1)
  train <- dataSet[ splitIndex,]
  test  <- dataSet[-splitIndex,]
  
  
  if(OU == "over") {
    train0 = train[train$target == '0',]
    train1 = train[train$target == '1',]
    n0 = nrow(train0)
    n1 = nrow(train1)

    train11 = train1[sample(1:n1, n0, replace = TRUE),]

    train_over <- rbind(train11, train0)

    return (train_over)
  } else if (OU == "under") {
    train0 = train[train$target == '0',]
    train1 = train[train$target == '1',]

    train00 = train0[sample(1:nrow(train0), nrow(train1)),]
    train_under <- rbind(train00, train1)

    return (train_under)
  } else {
    return (cat("Please enter \"over\" or \"under\" as your second argument"))
  }
}
Quick_bal(hmeq, "over")
Quick_bal(hmeq, "under")
```

```r
Quick_model_bal = function(dataSet) {
    splitIndex <- createDataPartition(dataSet$target, p = 0.70, list = FALSE, times = 1)
    train <- dataSet[ splitIndex,]
    test  <- dataSet[-splitIndex,]  
  
    train0 = train[train$target == '0',]
    train1 = train[train$target == '1',]
    n0 = nrow(train0)
    n1 = nrow(train1)

    train11 = train1[sample(1:n1, n0, replace = TRUE),]

    train_over <- rbind(train11, train0)

    model = ranger(target ~., data = train_over)
    pred = predict(model, data = test)$predictions
    cm = confusionMatrix(pred, test$target, positive = "1")

    accuracy11 = cm[[3]][1]
    balancedAccuracy11 = cm[[4]][11]

    train0 = train[train$target == '0',]
    train1 = train[train$target == '1',]

    train00 = train0[sample(1:nrow(train0), nrow(train1)),]
    train_under <- rbind(train00, train1)

    model = ranger(target ~., data = train_under)
    pred = predict(model, data = test)$predictions
    cm = confusionMatrix(pred, test$target, positive = "1")

    accuracy00 = cm[[3]][1]
    balancedAccuracy00 = cm[[4]][11]

    over_acc = balancedAccuracy11
    under_acc = balancedAccuracy00
    accuracies = rbind(over_acc, under_acc)
    return (accuracies)
}
Quick_model_bal(hmeq)
```
