```r
titanic=read.csv("C:\\Users\\student\\Documents\\Applied Data Mining\\train.csv")
library(ggplot2)
```

```r
funcOne = function(x) {
  rt = x^2 + 3*x + 1
  return (rt)
}
funcOne(2)
```

```r
funcTwo = function(x){
  
  sum = 0
  for (i in 1:length(x)) {
    if(sum == 0) {
        sum = x[i]
    } else {
        sum = sum * x[i]
    }
  }
  return (sum)
}
funcTwo(c(2,3,1,6))
```

```r
funcThree = function(x,y){
  for(i in 1:length(x)){
    if (is.na(x[i])) {
      x[i] = y
    }
  }
  return (x)
}
funcThree(c(1,4,2,NA,5,NA),2)
```

```r
funcFour = function(x) {
  vectorMean = mean(x, na.rm=TRUE)
  for(i in 1:length(x)) {
    if (is.na(x[i])) {
      x[i] = vectorMean
    }
  }
  return (x)
}
funcFour(c(1,2,3,NA,NA))
```

```r
funcFive = function(x) {
  for(i in 1:ncol(x)) {
    if(is.numeric(x[,i])){
      colMean = mean(x[,i],na.rm=TRUE)
      for(j in 1:nrow(x)) {
        if(is.na(x[j,i])) {
          x[j,i] = colMean
        }
      }
    }
  }
  return (x)
}
funcFive(titanic)
```

```r
funcSix = function(dataset, xPlot) {
  ggplot(dataset) + geom_bar(mapping=aes(x=xPlot),position="dodge")
}
funcSix(titanic, titanic$Survived)
```

```r
funcSeven = function(dataset, xPlot, yPlot) {
  ggplot(dataset) + geom_bar(mapping=aes(x=xPlot, fill=yPlot),position="dodge")
}
funcSeven(titanic, titanic$Survived, titanic$Sex)
```
