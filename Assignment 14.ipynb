```r
quick_visual = function(dataSet, type) {
  library (ggplot2)

  if(is.data.frame(dataSet) && type == 1) {
    data_factors <- dataSet[,sapply(dataSet, is.factor)]

    cat.5 = function(x) {
      l = (length(unique(x)) <= 5)
      return(l)
    }

    data_factors <- data_factors[, sapply(data_factors, cat.5)]

    combinations <- combn(names(data_factors), 2, simplify = TRUE)

    bar.charts <- function(d) {
      chart <- ggplot(d) +
        geom_bar(mapping = aes(x = d[,1], fill = d[,2]), position = "dodge") +
        xlab(names(d[1])) +
        labs(fill = names(d[2]))

      return(chart)
    }

    c <- function (x) {
      bar.charts(data_factors[,x])
    }

    charts <- apply(combinations, 2, c)

    return (charts)
  } else if (is.data.frame(dataSet) && type == 2) {

    for(i in 1:(ncol(dataSet))) {
      for (j in (j):ncol(x)) {
        if(is.factor(dataSet[,i]) && is.numeric(dataSet[,j]) && (nlevels(dataSet[,i]) < 5)) {
          print(ggplot(dataSet) + geom_density(mapping = aes(x = dataSet[,i]), na.rm = TRUE) +
                  facet_wrap(dataSet[,j]) + labs(x = names(dataSet)[i], title = names(dataSet)[i]))
        }
      }
    }

  } else if (is.data.frame(dataSet) && type == 3) {
    data_num <- dataSet[,sapply(dataSet, is.numeric)]
    nums <- unlist(lapply(dataSet, is.numeric))
    xnum = dataSet[ ,nums]
    pairs(xnum[,1:ncol(xnum)], lower.panel = NULL)
  }
}
```

```r
quick_model = function(dataSet){
  library(lattice)
  library(caret)

  splitIndex <- createDataPartition(dataSet$target, p = .70, list = FALSE, times = 1)
  train <- x[splitIndex,]
  test <- x[-splitIndex,]

  library(ranger)

  model = ranger(target ~., data = train)
  pred  = predict(model, data = test)$predictions
  cm=confusionMatrix(pred, test$target, positive=NULL)

  return(cm$overall['Accuracy'])
}
```
